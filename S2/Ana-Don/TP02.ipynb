```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1-

```python
Z = np.array([
    [14,13,1.50,45],
    [16,13,1.60,50],
    [15,13,1.65,50],
    [9,15,1.75,60],
    [10,14,1.70,60],
    [7,14,1.70,60],
    [8,14,1.60,70],
    [13,13,1.60,65],
    [17,15,1.55,60],
    [11,14,1.70,65]
])

# Normalisation des données
Z = (Z - Z.mean(axis=0)) / Z.std(axis=0)

print("-----------------------------------------------------------------------------")
print(Z)
print("-----------------------------------------------------------------------------")

# Matrice de covariance

X = np.corrcoef(Z.T)

print(X)
print("-----------------------------------------------------------------------------")

# Valeurs propres et vecteurs propres de la matrice de covariance

eigval, eigvec = np.linalg.eig(X)

print(eigval)
print("-----------------------------------------------------------------------------")

# Tri des valeurs propres (car on veut les plus grandes valeurs car il ont le plus d'inertie)

eigval = sorted(eigval, reverse=True)

print(eigval)
print("-----------------------------------------------------------------------------")

# Inertie de chaque composante principale 

eig_inertie = eigval / np.sum(eigval)

print(eig_inertie)
print("-----------------------------------------------------------------------------")

# Inertie cumulée

eig_cum = np.cumsum(eig_inertie)

print(eig_cum)
print("-----------------------------------------------------------------------------")

```

## 2-

```python

# Affichage des résultats

mat = np.hstack((np.hstack((np.array(eigval).reshape(-1,1), eig_inertie.reshape(-1,1))), eig_cum.reshape(-1,1)))
print(mat)

```
