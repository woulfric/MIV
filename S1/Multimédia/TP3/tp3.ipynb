```python
def dct(u,x,f):
    val = math.cos(((2*x+1)*u*math.pi)/(2*len(f)))
    return val
```

```python
def alpha(u):
    return (1/math.sqrt(2)) if u==0 else 1
```

DCTU

```python
def DCT_u(u,f):
    # N = len(f)
    val = math.sqrt(2)/math.sqrt(len(f))*alpha(u)*sum([dct(u,x,f)*f[x] for x in range(len(f))])
    return val
```

DCTU Pour une ligne

```python
for i in range(len(image[0])):
    print(DCT_u(i,image[0]))
```

DCTU Pour une matrice

```python
img = image

for j in range(len(img)):
    print("\nColone numero :",j+1)
    for i in range(len(img[j])):
        print(DCT_u(i,img[j]))
```

DCT Inverse

```python
def DCT_invert(x, f):
    # N = len(f)
    val = math.sqrt(2)/math.sqrt(len(f))*sum([alpha(u)*dct(u,x,f)*DCT_u(u,f) for u in range(len(f))])
    return val
```

DCT inverse pour une ligne

```python
img = image1

for i in range(len(img[j])):
    print(DCT_invert(i,img[j]))
```

DCT Inverse pour une matrice

```python
img = image

for j in range(len(img)):
    print("Ligne numero : ",j+1)
    for i in range(len(img[j])):
        print(DCT_invert(i,img[j]))
```
