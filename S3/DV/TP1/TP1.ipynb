```python
import pandas as pd
import matplotlib.pyplot as plt
```

Exo 1

1-


```python
df_productivity = pd.read_csv("us-productivity.csv")

plt.figure()
plt.bar(df_productivity['state'],df_productivity['value'])
plt.xlabel("state")
plt.ylabel("value")
plt.xticks(rotation =90)
plt.show()
```

2-

```python
df_cities = pd.read_csv("us-cities.csv")
df_cities
```

```python
plt.figure(figsize= (20,20))
plt.barh(df_cities['place'],df_cities['population'])
plt.xlabel("Population")
plt.ylabel("place")
plt.xticks(rotation =90)
plt.show()
```

Exo 2

1-

```python
plt.figure(figsize=(10,8))
plt.scatter(df_cities['lon'],df_cities['lat'], c= df_cities['population'], s =[pop/1000 for pop in df_cities["population"]])
for i, place in enumerate(df_cities['place']):
    plt.text(df_cities['lon'][i], df_cities['lat'][i], place, fontsize = 6, ha = 'center', va = 'bottom')
plt.colorbar(label = "Population")
plt.xlabel("Longitude")
plt.xlabel("Latittude")
plt.show()
```
